- дальнейшей выгод от смены тарифа, 
- выработки предложений по переходу каждого клиента на более выгодный и оптимальный тариф, 
- изучения возможного оттока клиентов.

**Задачи исследования**:

- исследовательский анализ данных

По результатам анализа необходимо разработать рекомендации о переходе отельных клиентов на более оптимальный тариф, а также проверить 2 дополнительных статистических гипотезы.

### Замена типов и замена пропусков

#### Даты

Судя по всему, в наборе данных dataset дата записана как строка в формате ГГГГ-ММ-ДД 00:00:00, т.е. время в данном атрибуте не несёт никакой смысловой нагрузки, поэтому от него можно избавиться

```python
clients['date_start'] = pd.to_datetime(clients['date_start'], format = '%Y-%m-%d').dt.date
dataset['date'] = pd.to_datetime(dataset['date'], format = '%Y-%m-%d').dt.date
clients['date_start'] = clients['date_start'].astype('datetime64[D]')
dataset['date'] = dataset['date'].astype('datetime64[D]')
```

```python
print('Период наблюдений в dataset c ', dataset['date'].min().date(), 'по', dataset['date'].max().date())
print('Период наблюдений в clients c ', clients['date_start'].min().date(), 'по', clients['date_start'].max().date())
```

#### ID оператора

Без предварительной замены пропусков в данном атрибуте просто так на тип int заменить не возможно. В целом можно применить преобразование в Int64. Проверим, как распеределены пропуски относительно других атрибутов

```python
for i in dataset.columns.drop(['user_id', 'operator_id']):
    display(dataset.query('operator_id.isna()').groupby(i)['user_id'].count().reset_index().sort_values(by='user_id', ascending=False).head(20))
```

В основном пропуски в атрибуте с информацией об операторе характерны для **внутренних входящих пропущенных звонков** (следовательно, длительностью 0). Есть 52 клиента, у которых колиество внутренних звонков с отсутсвущей информацией об операторе, превышает 50, всего таких клиентов 305. Т.е. почти для половины клиентов в наборе данных clients отсутсвует информация об операторе для части звонков. Таким образом полноценной замены пропусков (для каждого клиента и типа звонка в зависимости от категориального атрибута задать соответствующий номер оператора) в значениях атрибута operator_id выполнить невозмоно. 

Заменим пропуски на некоторе число. По всей видимости, номер оператора - число с 6-ю разрядами. Пусть неизвестный оператор будет под номером 1000000.

Как выяснится, пропущенный оператор етсь практически у каждого клиента.

```python
#проверим срез
dataset_query = dataset.query('operator_id.isna() & direction == "in"').groupby('user_id')['calls_count'].count().reset_index().sort_values(by='calls_count', ascending=False)
print('Количество клиентов, для звонков которых отсутсвует информация об операторе:', dataset_query.shape[0])
print('Количество клинетов, информация о звонках которых есть в наборе данных dataset', dataset['user_id'].nunique())
print('Количество клиентов, количество звонков которых с отсутсвующией информацие об операторах превышает 50:', dataset_query.query('calls_count > 50').shape[0])
```

#### Ед. измерения. Минуты или секунды



Т.к. информация о тарифных планах и их расценки представлены в минутах за звонок, то логично предположить, что длительность звонков в атрибутах call_duration и total_call_duration измеряется в минутах.Если бы измерение велось, например, в часах, то в данных атрибутах мы бы выявили значения типа float.

С другой стороны, 166155 минут на одного оператора в день (или среднее значение 1158 минут) в часах - это: 

```python
print(166155 / 60, 'часов и', 1158 / 60, 'часов соответственно')
```

Проверим, чему равна медиана длительности звонка (без задержки и с задержкой) в зависимости от дня недели, скорее всего по выходным она минимальна. Для этого создадим доп. столбец с днем недели, и посчитаем медину длительности по дате. Необходимо будет сделать двойную группировку.

```python
dataset['weekday'] = dataset['date'].dt.weekday
dataset['one_call_duration'] = dataset['call_duration'] / dataset['calls_count']
dataset['total_one_call_duration'] = dataset['total_call_duration'] / dataset['calls_count']
display(dataset.groupby(['date','weekday'])['one_call_duration'].median().reset_index().groupby('weekday')['one_call_duration'].median().reset_index())
display(dataset.groupby(['date','weekday'])['total_one_call_duration'].mean().reset_index().groupby('weekday')['total_one_call_duration'].mean().reset_index())
#display(dataset.groupby('weekday')['one_call_duration'].median().reset_index())
```

Пока не ясно. 20 секунд разговора? Не совсем реалистично, при этом мы еще не удалили часть аномально высоких значений в длительностях звонков. С учетом ожидания - 70-80 секунд?

Хотя мы не знаем специфику и направление деятельности клиентов провайдера. Мб, клиенты - это муниципальные службы, которые принимают звонки от своих контрагентов по определенному скрипту: "Сеточка, прими информацию, столовая №15, 20 сентября, 3 пирожка, 20 котлет, 6 компотов", т.е. разговор в целом укладывается в 20 секунд и не длится больше минуты.

Посчитаем также медианное значение общей длительности всех звонков, приходящееся на одного оператора в день. Так мы сможем проверить ед. измерения через длительность рабочего дня 

```python
# сначала просумируем длительность всех звонков за каждый день по всем операторам,
# далее найдем медиану для каждого дня недели дл каждого оператора,
# далее найдем медиану для всех операторов по дням недели
dataset.groupby(['date', 'weekday', 'operator_id'])['total_call_duration'].sum().reset_index() \
                .groupby(['weekday', 'operator_id'])['total_call_duration'].median().reset_index() \
                .groupby('weekday')['total_call_duration'].median().reset_index()
```

Т.е. в будний день 1 оператор занят разговором, дозвоном и соединением с клиентами примерно 600 временных единиц. Если это секунды, то дело - труба, все очень неэффективно. Зачем столько операторов, которые работаю по 10 минут в день. Если это минуты, то уже реалистичнее. 720 минут - это почти  12 часовой рабочий день оператора колл-центра

### Переход на другие тарифы активных пользователей

**Описание методического подхода к опередлению оптимального тарифа для активного пользователя**

1) Активным пользователь считается, если он (клиент) пользуется услугами провайдера на протяжении 3-х месяцев и совершал звонков хотя бы больше 25 персентиля

2) Выбор оптимального тарифа происходит следующим образом:

- для каждого пользователя рассчитывается стоимость услуг на каждом тарифе исходя из объема услуг, который пользователь потребил за 3 месяца с сентября по ноябрь
- рассчитывается разница между расходами (сравниваются расходы) на текущем тарифе и на других тарифах
- оптимальным тариф считается тот, для которого разница расходов максимальна по сравнению с текущим тарифом

- пользователь в среднем должен звонить хотя бы больше значения 25 квартиля
 
 и соберем информацию по ним.

#### Предложения по переходу на оптимальные тарифы для активных пользователей

Создадим функцию, которая будет принимать на вход таблицу из предыдущего пункта, а именно столбцы со значениями разнцы расходов на текущем тарифе и другими тарифами при одинаковом потреблении услуг и возвращать наименование тарифа, разницу между расходами и общие расходы на тарифе. Т.е по сути - функция вернет минильное значение расходов для клиента с учетом потребления услуг в текущем месяце

```python
def rec_tariff(row):
    """
    Возвращает наименование тарифа, на который необходимо перейти, чтобы минимизировать расход на услуги,
    а также значение расходов на рекомендуемом тарифе
    
    """
    exp_a = row['exp_a']
    exp_b = row['exp_b']
    exp_c = row['exp_c']
    cur = row['exp_current']
    min_diff = min(exp_a, exp_b, exp_c)
    
    if min_diff == exp_a:
        rec = 'A'
        rec_exp = row['exp_a']
        rec_delta =  row['exp_current'] - row['exp_a']
    elif min_diff == exp_b:
        rec = 'B'
        rec_exp = row['exp_b']
        rec_delta = row['exp_current'] - row['exp_b']
    elif min_diff == exp_c:
        rec = 'C'
        rec_exp = row['exp_c']
        rec_delta = row['exp_current'] - row['exp_c']
    
    return pd.Series([rec, rec_exp, rec_delta])
```

Допустим, что в декабре активность клиентов соответствует активности в октябре-ноябре. Поэтому необходимо подключить какое-то число новых клиентов, которрое должно быть пропорционально их общим расходам на каждом тарифе 

```python
falling_exp = user_expenses.pivot_table(index=['tariff_plan'], columns='month', values=['user_id', 'expenses'],
                                            aggfunc={'user_id' : 'nunique', 'expenses' : 'sum'}).reset_index()
falling_exp.columns = ['_'.join(map(str, col)) for col in falling_exp.columns]
#доли расходов относительно всех расходов
falling_exp['ratio_exp'] = (falling_exp['expenses_10'] / falling_exp['expenses_10'].sum() + falling_exp['expenses_11'] / falling_exp['expenses_11'].sum()) / 2
#средние расходы одного клиента
falling_exp['mean_user'] = falling_exp['expenses_10'] / falling_exp['user_id_10'] + falling_exp['expenses_11'] / falling_exp['user_id_11']
# распределение возмещаемых убытков 
falling_exp['new_exp'] = falling_exp['ratio_exp']*total_exp*1000000
#  количество новых пользователей
falling_exp['new_user'] = round(falling_exp['new_exp'] / falling_exp['mean_user'])

falling_exp[['tariff_plan_', 'new_user']]
```

#### Проверка данных на нормальность

Проведем проверку данных на нормальность, воспользуемся критерием Шапиро-Уилка

```python
# зададим список наших наборов данных
serieses = [(duration_out_active, 'duration_out_active'),\
          (duration_out_non_active, 'duration_out_non_active'),\
          (mean_active_exp, 'mean_active_exp'),\
          (mean_non_active_exp, 'mean_non_active_exp'),
          (duration_A, 'duration_A'),\
          (duration_B, 'duration_B'),\
          (duration_C, 'duration_C'),\
          (exp_A, 'exp_A'), (exp_B, 'exp_B'), (exp_C, 'exp_C'),\
          (exp_current, 'exp_current'), (exp_rec, 'exp_rec')]

# с помощью цикла преобразуем столбцы в списки и применим критерий Шапиро-Уилка, выведем результат
for i,j in serieses:
    lst = i.tolist()
    result = stats.shapiro(lst)
    p_value = result[1]
    print(f'р-значение для {j}:',p_value)
```

Все р-значения меньше α или равны 0, что свидетельствует о том, что данные распеределены ненормально. Это говорит от том, что скорее всего данные реальные.

Для проверки гипотез о равенстве 2-х средних будем применять Т-тест, хотя для ненормальных распеределний и выборок он вроде бы как и не применим, поэтому в дополнение к нему будем использовать непараметрический критерий Манна-Уитни и критерий Колмогорова-Смирнова. Данные 2  критерия позволят сравнить массивы разного размера, в отличие, например, от критерия Уилкоксона.

Также в целом применим тест Краскела-Уоллиса (Н-критерий) для проверки равенства медиан нескольких выборок

Основные особенности теста Манна-Уитни:
 - не требуется нормальное распределение - подходит
 - данные должны быть независимы - неизвестно.

Также в рамках проверки гипотез предполагается сравение более 2-х групп - средних по тарифам - суммарно 3 проверки, что может дать возможность ошибиться с Т-тестом с вероятностью 1 − (1- 0,05 )^3 = 0,142 или 14% при интерпретации результатов.

#### Функции расчета значений критериев

Создадим функции расчета значений критерия и проверки верности гипотез

```python
def mannwhitneyu(data_A, data_B):
    p_value = stats.mannwhitneyu(data_A, data_B, use_continuity=True, alternative='two-sided')[1]
    print('P-value: {0:.3f}'.format(p_value))
    print('Среднее первого набора данных: {0:.3f}'.format(data_A.mean()))
    print('Среднее второго набора данных: {0:.3f}'.format(data_B.mean()))
    if (p_value < α):
        print('Отвергаем H0: разница статистически значима')
    else:
        print('Не отвергаем H0: вывод о различии сделать нельзя')
```

```python
def ttest(data_A, data_B):
    p_value = stats.ttest_ind(data_A, data_B)[1]
    print('P-value: {0:.3f}'.format(p_value))
#    print('Среднее первого набора данных: {0:.3f}'.format(data_A.mean()))
#    print('Среднее второго набора данных: {0:.3f}'.format(data_B.mean()))
    if (p_value < α):
        print('Отвергаем H0: разница статистически значима')
    else:
        print('Не отвергаем H0: вывод о различии сделать нельзя')
```

```python
def kolmohorov(data_A, data_B):
    p_value = stats.ks_2samp(data_A, data_B)[1]
    print('P-value: {0:.3f}'.format(p_value))
#    print('Среднее первого набора данных: {0:.3f}'.format(data_A.mean()))
#    print('Среднее второго набора данных: {0:.3f}'.format(data_B.mean()))
    if (p_value < α):
        print('Отвергаем H0: разница статистически значима')
    else:
        print('Не отвергаем H0: вывод о различии сделать нельзя')
```

```python
def hyp(data_A, data_B):
    """
    Функция расчитывает значения p-value для трех тестов и выводит значения p-value, а также результат подтверждения
    или опроверженя гипотезы
    """
    
    pv_mw = stats.mannwhitneyu(data_A, data_B, use_continuity=True, alternative='two-sided')[1]
    pv_tt = stats.ttest_ind(data_A, data_B, equal_var=False)[1]
    pv_ks = stats.ks_2samp(data_A, data_B)[1]
    
    print('Среднее первого набора данных: {0:.2f}'.format(data_A.mean()))
    print('Среднее второго набора данных: {0:.2f}'.format(data_B.mean()))
    print('')
    
    print('P-value Т-теста: {0:.5f}'.format(pv_tt))
    if (pv_tt < α):
        print('Отвергаем H0: разница статистически значима')
    else:
        print('Не отвергаем H0: вывод о различии сделать нельзя')
    print('')
    
    print('P-value теста Манна-Уитни: {0:.5f}'.format(pv_mw))
    if (pv_mw < α):
        print('Отвергаем H0: разница статистически значима')
    else:
        print('Не отвергаем H0: вывод о различии сделать нельзя')
    print('')
    
    print('P-value теста Колмогорова-Смирнова: {0:.5f}'.format(pv_ks))
    if (pv_ks < α):
        print('Отвергаем H0: разница статистически значима')
    else:
        print('Не отвергаем H0: вывод о различии сделать нельзя')
          
          
```

А по идее можно записать все результаты сразу в таблицу

<div class="alert alert-info"> <b>Добавлен еще Н-тест и расчет медиан. Изменены формулировки вывода </div>

```python
def grand_hyp(hypothesis, data_pair):
    """
    Функция принимает на вход список пары данных в виде кортежей для проверки каждой гипотезы и возвращает p-значения тестов
    и результат проверки гипотезы
    """
    
    results = []
    
    for i, (hypothesis, data_pair) in enumerate(zip(hypothesis, data_pair), start=1):
        data_A, data_B = data_pair
        mean_A = data_A.mean()
        mean_B = data_B.mean()
        median_A = data_A.median()
        median_B = data_B.median()
        
        pv_tt= stats.ttest_ind(data_A, data_B, equal_var=False)[1]
        pv_mw = stats.mannwhitneyu(data_A, data_B, use_continuity=True, alternative='two-sided')[1]
        pv_ks = stats.ks_2samp(data_A, data_B, method='asymp')[1]
        pv_h = stats.kruskal(data_A, data_B)[1]
        
        if pv_tt < α:
            decision_tt = 'Отвергаем H0'
        else:
            decision_tt = 'Не отвергаем H0'
        
        if pv_mw < α:
            decision_mw = 'Отвергаем H0'
        else:
            decision_mw = 'Не отвергаем H0'
        
        if pv_ks < α:
            decision_ks = 'Отвергаем H0'
        else:
            decision_ks = 'Не отвергаем H0'
            
        if pv_h < α:
            decision_h = 'Отвергаем H0'
        else:
            decision_h = 'Не отвергаем H0'
        
        results.append([hypothesis, mean_A, mean_B, median_A, median_B, pv_tt, pv_mw, pv_ks, pv_h, decision_tt, decision_mw, decision_ks, decision_h])
    
    columns = ['Гипотеза №', 'Среднее 1', 'Среднее 2', 'Медиана 1', 'Медиана 2',
               'p-value (t-test)', 'p-value (Mann-Whitney U)', 'p-value (KS-test)', 'p-value (Н-тест)',
               'Решение (t-test)', 'Решение (Mann-Whitney U)', 'Решение (KS-test)', 'Решение (Н-тест)']
    result_data = pd.DataFrame(results, columns=columns)
    
    return result_data
```
